```python
import pandas as pd
import numpy as np
```

```python
Data = pd.read_excel("/Ola July Data.xlsx")
```

```python
Data.columns
```

```python
Checking_Data = Data.head
print(Checking_Data)
```

```python
# Retrieve all successful bookings
Successful_Bookings = Data[Data['Booking_Status'] =="Success"]

#Count of Sucessfull Booking
Total_Count = len(Successful_Bookings)
print('Total_Successful_Bookings :',Total_Count)
```

```python
#Find the average ride distance for each vehicle type
Average_Ride_Distance = Data.groupby('Vehicle_Type')['Ride_Distance'].mean()
print(Average_Ride_Distance)
```

```python
# Get the total number of cancelled rides by customers
Cancelled_Rides_by_Customer = Data[Data['Booking_Status'] == 'Canceled by Customer']
Total_Cancelled_Rides = len(Cancelled_Rides_by_Customer)
print('Total_Cancelled_Rides :',Total_Cancelled_Rides)
```

```python
# List the top 5 customers who booked the highest number of rides
Top_5_Customers = Data.groupby('Customer_ID')['Booking_Value'].sum().sort_values(ascending=False).head(5)
print(Top_5_Customers)
```

```python
# Get the number of rides cancelled by drivers due to personal and car-related issues
Cancelled_Rides_by_Drivers = Data[Data['Canceled_Rides_by_Driver'] == 'Personal & Car related issue']
Total_Cancelled_Rides = len(Cancelled_Rides_by_Drivers)
print('Total_Cancelled_Rides :', Total_Cancelled_Rides)
```

```python
# Find the maximum and minimum driver ratings for Prime Sedan bookings:
Prime_Sedan_Ratings = Data[Data['Vehicle_Type'] == 'Prime Sedan']
Max_Rating = Prime_Sedan_Ratings['Driver_Ratings'].max()
Min_Rating = Prime_Sedan_Ratings['Driver_Ratings'].min()
print("Maximum Rating:", Max_Rating)
print("Minimum Rating:", Min_Rating)
```

```python
# Retrieve all rides where payment was made using UPI:
UPI_Payments = Data[Data['Payment_Method'] == 'UPI']
print(UPI_Payments)
```

```python
# To display in Table format.
from IPython.display import display
display(UPI_Payments)
```

```python
# Count of total payment done by UPI.
Total_UPI_Count = UPI_Payments.shape[0]
print('Total_UPI_Count :', Total_UPI_Count)
```

```python
# Find the average customer rating per vehicle type:
Average_Customer_Rating = Data.groupby('Vehicle_Type')['Customer_Rating'].mean().round(2)
print(Average_Customer_Rating)
```

```python
# Calculate the total booking value of rides completed successfully:
Successful_Rides= Data[Data['Booking_Status'] == 'Success']
Total_Booking_Value = Successful_Rides['Booking_Value'].sum()
print('Total Booking Value of Successful Rides :', Total_Booking_Value)
```

```python
# List all incomplete rides along with the reason:
Incomplete_Rides = Data[Data['Incomplete_Rides'] == 'Yes']['Incomplete_Rides_Reason'].value_counts()
print('Incomplete Rides with reason are : ',Incomplete_Rides)
```
